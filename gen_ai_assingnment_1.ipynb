```python
1. Difference between "Love" and "love" in NLP
In Natural Language Processing (NLP), "Love" and "love" are treated as different words if case sensitivity is not handled. "Love" may appear at the beginning of a sentence or as a proper noun, while "love" is commonly used as a verb or noun representing emotion. This increases vocabulary size and creates inconsistency. To solve this, we apply lowercasing so both are converted into "love".

2. What happens if stopwords are not removed?
If stopwords such as "is", "the", and "and" are not removed, the text becomes noisy and less meaningful. It can reduce model accuracy, increase processing time, and make it harder to identify important words. Removing stopwords helps focus on meaningful content.

3. Two real-world scenarios where removing stopwords can be harmful
a) Sentiment Analysis:
In the sentence "This is not good", removing the word "not" changes the meaning to "good", which is incorrect.
b) Question Answering Systems:
In the sentence "What is the capital of India?", removing stopwords results in "capital India", which loses the structure and meaning of the question.

4. Difference between Stemming and Lemmatization
Stemming reduces words to their root form by removing suffixes, but the result may not be a real word (e.g., "running" → "runn"). Lemmatization converts words to their base dictionary form using grammar rules (e.g., "running" → "run"). Stemming is faster but less accurate, while lemmatization is more accurate and meaningful.
```

```python
import re

def preprocess_text(text):
    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Handle repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. Remove punctuation (keep only words and spaces)
    text = re.sub(r'[^\w\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Remove very short tokens (≤ 2), keep "no", "not"
    tokens = text.split()
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    return " ".join(cleaned_tokens)


# ✅ Example Test Cases
texts = [
    "I have 2 dogs",
    "This    is    good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now",
    "Contact me at test@gmail.com",
    "This is not bad at all"
]

for t in texts:
    print("Original:", t)
    print("Processed:", preprocess_text(t))
    print("-" * 50)
```

```python
# Sample Inputs
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Stress Testing
for i, sentence in enumerate(test_sentences, 1):
    cleaned_sentence = preprocess_text(sentence)
    tokens = cleaned_sentence.split()

    print(f"--- Test Case {i} ---")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print()
```

```python
# Function to compute analytics
def token_analytics(sentence):
    tokens = sentence.split()

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens if total_tokens > 0 else 0

    return total_tokens, unique_tokens, round(avg_length, 2)


# Run analytics on test sentences
for i, sentence in enumerate(test_sentences, 1):
    cleaned_sentence = preprocess_text(sentence)
    total, unique, avg_len = token_analytics(cleaned_sentence)

    print(f"--- Test Case {i} ---")
    print("Original Text      :", sentence)
    print("Cleaned Sentence   :", cleaned_sentence)
    print("Total Tokens       :", total)
    print("Unique Tokens      :", unique)
    print("Avg Token Length   :", avg_len)
    print()
```

```python
from collections import Counter

# Collect all tokens from all sentences
all_tokens = []

for sentence in test_sentences:
    cleaned_sentence = preprocess_text(sentence)
    tokens = cleaned_sentence.split()
    all_tokens.extend(tokens)

# Count frequency
word_counts = Counter(all_tokens)

# Top 10 most frequent words
top_10 = word_counts.most_common(10)

# Top 5 least frequent words
# (words with lowest frequency)
least_5 = sorted(word_counts.items(), key=lambda x: x[1])[:5]

print("Top 10 Most Frequent Words:")
for word, freq in top_10:
    print(f"{word}: {freq}")

print("\nTop 5 Least Frequent Words:")
for word, freq in least_5:
    print(f"{word}: {freq}")
```

```python
import re

def preprocess_text(text):
    # ✅ Handle empty or None input
    if not text or not isinstance(text, str):
        return ""

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. Remove emojis (basic range)
    text = re.sub(r'[^\w\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # ✅ Handle case where nothing remains
    if text == "":
        return ""

    # 8. Remove very short tokens (≤ 2), keep "no", "not"
    tokens = text.split()
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    return " ".join(cleaned_tokens)
```

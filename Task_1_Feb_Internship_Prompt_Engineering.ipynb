```python
# Teaching Template
teaching_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# Storytelling Template
story_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)
```

```python
def prompt_engine(topic):
    teaching = teaching_prompt.format(topic=topic)
    interview = interview_prompt.format(topic=topic)
    storytelling = story_prompt.format(topic=topic)

    return teaching, interview, storytelling
```

```python
topic = "Machine Learning"

teaching, interview, storytelling = prompt_engine(topic)

print("Teaching Prompt:\n", teaching)
print("\nInterview Prompt:\n", interview)
print("\nStorytelling Prompt:\n", storytelling)
```

```python
templates = {
    "teaching": teaching_prompt,
    "interview": interview_prompt,
    "story": story_prompt
}

def get_prompt(template_type, topic):
    if template_type not in templates:
        return "Invalid template type"

    return templates[template_type].format(topic=topic)

print(get_prompt("teaching", "Machine Learning"))
```

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        raise ValueError("Invalid audience! Choose from beginner, intermediate, expert")

    if tone not in valid_tone:
        raise ValueError("Invalid tone! Choose from formal, casual, fun")

    return audience, tone
```

```python
templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style"
    ),

    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Ask interview questions about {topic} for {audience} in a {tone} tone"
    ),

    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style"
    )
}
```

```python
def generate_prompt(topic, audience, tone, style):

    # Validate style
    if style not in templates:
        return "Invalid style! Choose from teaching, interview, storytelling"

    # Generate prompt
    return templates[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
def generate_prompt(topic, audience, tone, style):

    valid_styles = ["teaching", "interview", "storytelling"]

    if style not in valid_styles:
        return "Invalid style!"

    if not topic or not audience or not tone:
        return "All inputs are required!"

    return templates[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)
```

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"
    if tone not in valid_tone:
        tone = "formal"

    return audience, tone
```

```python
def generate_prompt(topic, audience, tone):

    audience, tone = validate_inputs(audience, tone)

    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "casual"),
    ("Deep Learning", "expert", "formal"),
    ("Data Science", "beginner", "casual"),
    ("NLP", "expert", "fun")
]

for topic, audience, tone in inputs:
    print(generate_prompt(topic, audience, tone))
    print("-" * 50)
```

```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

pipe = pipeline(
    "text-generation",
    model="google/flan-t5-base",
    max_new_tokens=200
)

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain_core.prompts import PromptTemplate

extract_prompt = PromptTemplate.from_template("""
Extract skills, experience, and tools from:

{resume}
""")

match_prompt = PromptTemplate.from_template("""
Compare resume with job description:

Resume:
{resume_data}

Job:
{job_description}

Give matching and missing skills.
""")

score_prompt = PromptTemplate.from_template("""
Based on match:

{match_data}

Give score (0–100)
""")

explain_prompt = PromptTemplate.from_template("""
Explain the score:

Match:
{match_data}

Score:
{score}
""")
```

```python
extract_chain = extract_prompt | llm
match_chain = match_prompt | llm
score_chain = score_prompt | llm
explain_chain = explain_prompt | llm
```

```python
resumes = {
    "Strong": "Python developer with ML, NLP, TensorFlow, 3 years experience",
    "Average": "Python, SQL, basic ML, 2 years",
    "Weak": "Fresher, C, HTML"
}

job_description = "Looking for Data Scientist with Python, ML, NLP, SQL"
```

```python
for label, resume in resumes.items():
    print(f"\n===== {label} Candidate =====")

    extracted = extract_chain.invoke({"resume": resume})
    print("Extracted:", extracted)

    match = match_chain.invoke({
        "resume_data": str(extracted),
        "job_description": job_description
    })
    print("Match:", match)

    score = score_chain.invoke({"match_data": str(match)})
    print("Score:", score)

    explanation = explain_chain.invoke({
        "match_data": str(match),
        "score": str(score)
    })
    print("Explanation:", explanation)
```
